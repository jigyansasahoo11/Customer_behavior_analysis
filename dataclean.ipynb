```python
df=df.rename(columns={'purchase_amount_(usd)':'purchased_amount'})
df.columns
```

```python
labels=['young adult','adult','middle_age','senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
frequency_mapping = {
    'fortnightly':14,  
    'weekly':7,
    'monthly':30,
    'quarterly':90,  
    'bi-weekly':14,
    'annually':365,    
    'every 3 months':90
}
```

```python
df['purchases_of_frequency']=df['frequency_of_purchases'].str.lower().map(frequency_mapping)
df[['purchases_of_frequency','frequency_of_purchases']]
```
